```python
import pandas as pd
```

QUESTÃO 1 - Utilizando groupby(), calcule o total de vendas para cada produto.

```python
df = pd.DataFrame({
    "Produto": ["Notebook","Notebook","Mouse","Mouse"],
    "Vendas": [10,15,20,25]
})

vendas_produtos = df.groupby("Produto")["Vendas"].agg("sum")
print("Vendas por produto: ")
display(vendas_produtos)
```

QUESTÃO 2 - Calcule a média de vendas por cidade.

```python
df = pd.DataFrame({
    "Cidade": ["SP","SP","RJ","RJ"],
    "Vendas": [10,20,30,40]
})

media_cidades = df.groupby("Cidade")["Vendas"].agg("mean")
print("Media de vendas por cidade: ")
display(media_cidades)
```

QUESTÃO 3 - Conte quantas vezes cada produto aparece no DataFrame.

```python
df = pd.DataFrame({
    "Produto": ["Notebook","Mouse","Mouse","Teclado","Notebook"]
})

contagem_produtos = df["Produto"].value_counts()
print("Contagem dos produtos: ")
display(contagem_produtos)
```

QUESTÃO 4 - Agrupe os dados por Produto e Cidade e calcule a soma das vendas.

```python
df = pd.DataFrame({
    "Produto": ["Notebook","Notebook","Mouse","Mouse"],
    "Cidade": ["SP","RJ","SP","RJ"],
    "Vendas": [10,15,20,25]
})

print("Soma das vendas por produto e cidade:")
display(df.groupby(["Produto","Cidade"]).agg("sum"))
```

QUESTÃO 5 - Descubra qual foi a maior venda registrada para cada produto.

```python
df = pd.DataFrame({
    "Produto": ["Notebook","Notebook","Mouse","Mouse"],
    "Vendas": [10,50,20,25]
})

maiores_vendas = df.groupby("Produto")["Vendas"].agg("max")
print("Maior venda de cada produto:")
display(maiores_vendas)
```

QUESTÃO 6 - Calcule o preço médio de cada produto.

```python
df = pd.DataFrame({
    "Produto": ["Notebook","Notebook","Mouse","Mouse"],
    "Preco": [3000,3200,50,55]
})

precos_medios = df.groupby("Produto")["Preco"].agg("mean")
print("Preço médio por produto:")
display(precos_medios)
```

QUESTÃO 8 - Calcule o salário médio por departamento.

QUESTÃO 10 - Calcule o total de vendas por produto e ano.

QUESTÃO 12 - Calcule a temperatura máxima por cidade.

```python
df = pd.DataFrame({
    "Cidade": ["SP","SP","RJ","RJ"],
    "Temperatura": [30,32,28,27]
})

temperaturas_maximas = df.groupby("Cidade")["Temperatura"].agg("max")
print("Temperatura máxima por cidade:")
display(temperaturas_maximas)
```

QUESTÃO 13 - Calcule a menor quantidade vendida por produto.

```python
df = pd.DataFrame({
"Produto": ["Notebook","Notebook","Mouse","Mouse"],
"Quantidade": [5,2,10,8]
})

menores_quantidades_vendidas = df.groupby("Produto")["Quantidade"].agg("min")
print("Menor quantidade vendida por produto:")
print(menores_quantidades_vendidas)
```

QUESTÃO 14 - Calcule o preço médio e a soma da quantidade por categoria.

```python
df = pd.DataFrame({
    "Categoria": ["Eletronico","Eletronico","Movel","Movel"],
    "Preco": [1000,1200,500,700],
    "Quantidade": [2,3,1,4]
})

df14 = df.groupby("Categoria").agg({
    "Preco": "mean",
    "Quantidade": "sum"
})
df14.rename(columns={"Preco":"Preco medio"}, inplace=True)
print("Preço médio e soma das quantidades por categoria:")
display(df14)
```

QUESTÃO 15 - Agrupe os dados por região e ordene o resultado pelas maiores vendas.

```python
df = pd.DataFrame({
    "Regiao": ["Norte","Sul","Norte","Sudeste","Sul"],
    "Vendas": [100,200,150,300,250]
})

df_por_regiao = df.groupby("Regiao")["Vendas"].agg("sum")
df_por_regiao = df_por_regiao.sort_values(ascending=False)
print("Vendas por região(da maior para a menor):")
display(df_por_regiao)
```
