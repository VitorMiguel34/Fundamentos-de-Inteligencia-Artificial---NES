EXERCICIO 35 - Gere uma matriz 1000 × 1000 e calcule o tempo de multiplicação matricial.

```python
import time

arr35a = np.zeros(shape=(1000,1000))
arr35b = np.ones(shape=(1000,1000))

inicio = time.perf_counter()

mult = np.multiply(arr35a,arr35b)

fim = time.perf_counter()

tempo_execucao = fim - inicio

print(f"Multiplicacao realizada em {tempo_execucao} segundos")
```
